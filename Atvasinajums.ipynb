# Atvasinājums funkcijai (sinx^2)

1) No "matplotlib" importē "pyplot as plt" un no "numpy" importē "sin, linspace".

```python
from matplotlib import pyplot as plt
from numpy import sin, linspace
```

2) Mainīgo definēšana un to attēlojums.

```python
def f(x):
    return sin(x**2)

x = linspace(0, 4, 10)
print(x)
y = f(x)
print(y)
```

3) Mainīgo definēšana abiem atvasinājumiem ,un formulas izveide pirmajam atvasinājumam.

```python
firstder = [] 
deltax = x[1] - x[0]
N = len(x)

for i in range(N):
    temp = (f(x[i] + deltax) - f(x[i])) / deltax
    firstder.append(temp)
```

4) Formulas izveide otrajam atvasinājumam.

```python
secondder = [] 

for i in range(N - 1):
    temp = (firstder[i + 1] - firstder[i]) / deltax
    secondder.append(temp)

legend = []
```

5) Koda daļa, kurā tiek aprakstīts atvasinājumu vizuālizācija ar grafika palīdzību.

```python
legend = []
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")

plt.title("$sin(x^2)$ funkcija un tas atvasinājumi")

plt.plot(x, y, "m")
legend.append("$sin(x^2) funkcija$")
plt.plot(x, y, "go")
legend.append("$sin(x^2) funkcija(punkti)$")

plt.plot(x, firstder, "k")
legend.append("$sin(x^2)$ 1.k. atvasinājums")
plt.plot(x, firstder, "ro")
legend.append("$sin(x^2)$ 1.k. atvas.(punkti)")

plt.plot(x[0:len(x) - 1], secondder, "turquoise")
legend.append("$sin(x^2)$ 2.k. atvasinājums")
plt.plot(x[0:len(x) - 1], secondder, "ko")
legend.append("$sin(x^2)$ 2.k. atvas.(punkti)")

plt.legend(legend, loc=3) #Leģendas lokācija attēlā

plt.show() 
```
