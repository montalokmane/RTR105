# Integrēšana funkcijai sin( x^2)

Pirmais solis ir no "numpy" importēt "random,sin " un no "matplotlib" importēt "pyplot as plt".

```python
from numpy import random,sin 
from matplotlib import pyplot as plt
```

Otrais solis ir mainīgo definēšana.

```python
N = 10000

a = 0
b = 5
c = 1

x= random.uniform(a, b, N)
y = random.uniform(a, c, N)
```

Trešais solis ir aprakstīt koda daļu, kas vizualizē grafiski, to kas izdarīts kodā.

```python
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.title('Funkcija sin x^2 un tās integrālis')
plt.plot(x, y, 'ko')
```

Ceturtais solis ir koda matemātiskā daļa, kur tiek izveidota formula, kas aprēķina sin(x^2) integrāli, kā arī attēlo aprēķināto grafiski.

```python
N1 = 1

for i in range(N):
    if y[i] < (sin(x[i]*x[i] ))/ 2:
        plt.plot(x[i], y[i], 'go')
        N1 = N1 + 1
    else:
        plt.plot(x[i], y[i], 'ro')
        
plt.show()
```

Pēdējais solis ir laukumu aprēķināšana un iegūtā integrāļa attēlošana grafiski ar  gadījuma skaitļiem.

```python
S = (b - a) * (c - a)
jaunaS = float(S * N1 / N)
print("Laukuma skaitliskā vertība:", )
print(jaunaS)
```
